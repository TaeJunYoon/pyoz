```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-colorblind')
%matplotlib inline
import numpy as np

import pyoz as oz
```

```python
def dpd_func(r, a):
    cutoff = np.abs(r - 1.0).argmin()
    dpd = np.zeros_like(r)
    dpd[:cutoff] = 0.5 * a * (1 - r[:cutoff])**2
    return dpd

dpd_unary = oz.System()
r = dpd_unary.r
dpd_unary.set_interaction(0, 0, dpd_func(r, a=50))
U_r = dpd_unary.U_r
```

```python
plt.plot(r, U_r[0, 0])
plt.xlabel('r')
plt.ylabel('U(r)')
plt.xlim(0, 2)
```

```python
g_r, c_r, e_r, h_k = dpd_unary.solve(rhos=5, closure_name='hnc')
```

```python
plt.plot(r, g_r[0, 0])
plt.xlabel('r')
plt.ylabel('g(r)')
plt.xlim(0, 4)
```
