```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-colorblind')
import seaborn as sns

import pyoz as oz

sns.set_style('whitegrid', {'xtick.major.size': 5,
                            'xtick.labelsize': 'large',
                            'ytick.major.size': 5,
                            'ytick.labelsize': 'large',
                            'axes.edgecolor': 'k',
                            'font.weight': 'bold',
                            'axes.labelsize': 'large',
})
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)
plt.style.use('seaborn-colorblind')
```

```python
r = np.arange(0.05, 8192*0.05, 0.05)
```

```python
sig_c = 5
sig_d = 2 
rho_d = 0.5 
U1 = oz.soft_depletion(r, eps=1, sig_c=sig_c, sig_d=sig_d, n=36, rho_d=rho_d)
sig_d = 1 
U2 = oz.soft_depletion(r, eps=1, sig_c=sig_c, sig_d=sig_d, n=36, rho_d=rho_d)
sig_d = 2 
rho_d = 1 
U3 = oz.soft_depletion(r, eps=1, sig_c=sig_c, sig_d=sig_d, n=36, rho_d=rho_d)
```

```python
plt.plot(r, U1, label=r'$\sigma_{depl} = 2, \rho_{depl} = 0.5$')
plt.plot(r, U3, label=r'$\sigma_{depl} = 2, \rho_{depl} = 1.0$')
plt.plot(r, U2, label=r'$\sigma_{depl} = 1, \rho_{depl} = 0.5$')
plt.xlim(sig_c * 0.75, sig_c*2)
plt.ylim(min(x.min() for x in [U1, U2, U3]) - 0.5, 2)

plt.title(r'$\sigma_{coll}$ = 5')
plt.ylabel('U(r), energy', size=20)
plt.xlabel('r, distance', size=20)
plt.legend(loc='best')
plt.savefig('soft_depl_Ur.pdf', bbox_inches='tight')
```
